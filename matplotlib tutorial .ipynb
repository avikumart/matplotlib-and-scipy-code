```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("show the plot", fontdict={'color':'red','size':15})
plt.ylabel("calories burned")
plt.xlabel("avg pulse")
plt.grid(color="green", ls=':',lw=1)
plt.plot(x,y, marker='x', ms=5, linestyle=':')
plt.show
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y, color="red", width=0.5)
plt.title("bar chart",fontdict={'size':20})
plt.show()
```

```python
x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.title("histogram")
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0,0,0,0.3]

plt.pie(y, labels = mylabels, startangle=90, explode=myexplode)
plt.legend(title="fruits")
plt.show() 
```
